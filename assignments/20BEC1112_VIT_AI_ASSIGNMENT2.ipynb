```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("./drug200.csv")
```

```python
data1 = data
data1
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
data1['Sex'] = le.fit_transform(data1['Sex'])
data1['BP'] = le.fit_transform(data1['BP'])
data1['Cholesterol'] = le.fit_transform(data1['Cholesterol'])
data1['Drug'] = le.fit_transform(data1['Drug'])
```

```python
X = data.iloc[:,0:5]
```

```python
X
```

```python
Y = data['Drug']
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
Drug_model = Sequential()

Drug_model.add(Dense(5,activation='relu'))
Drug_model.add(Dense(64,activation='relu'))
Drug_model.add(Dense(128,activation='relu'))
Drug_model.add(Dense(32,activation='relu'))
Drug_model.add(Dense(8,activation='relu'))
Drug_model.add(Dense(1,activation='softmax'))
```

```python
Drug_model.compile(optimizer='adam', loss='mse')
```

```python
Drug_model.fit(X_train,Y_train,batch_size=4,epochs=100)
```

```python
Y_pred=Drug_model.predict(X_test)
```

```python
comparison = pd.DataFrame(Y_test)
```

```python
comparison.columns=['Actual Values']
comparison
```

```python
comparison['Predicted Values']=Y_pred
comparison
```

```python
#Custom Inputs
# Prepare the custom inputs
custom_inputs = np.array([[16, 0, 0, 0, 0.7]])  # Age, Sex, BP, Cholesterol, Na_to_K

# Make the prediction
predictions = Drug_model.predict(custom_inputs)

# Get the predicted drug class
predicted_class = np.argmax(predictions)

# Decode the predicted class back to the original label
predicted_drug = le.inverse_transform([predicted_class])

print(f"The predicted drug class is: {predicted_drug[0]}")
```
